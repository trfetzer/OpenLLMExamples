```python
from openai import OpenAI

# ---------------------------------------------------------------------------
# 1.  Set up the OpenAI client so it talks to your local Ollama daemon
# ---------------------------------------------------------------------------
client = OpenAI(
    base_url="http://0.0.0.0:11434/v1",  # or "http://localhost:11434/v1"
    api_key="ollama"                     # any nonâ€‘empty string is fine
)

# ---------------------------------------------------------------------------
# 2.  Craft the prompt
# ---------------------------------------------------------------------------
model = "phi4:latest"  # pick any model you have pulled with `ollama pull ...`

question = "What are the economic impacts of inflation?"

system_prompt = (
    "You are an AI assistant. Classify the following question into one of the "
    "predefined categories: 'Economics', 'Politics', 'Technology', or 'Health'. "
    "Return **only** JSON in the exact form {'category': '<category_name>'}."
)

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user",   "content": question}
]

# ---------------------------------------------------------------------------
# 3.  Call the chat endpoint
# ---------------------------------------------------------------------------
chat_completion = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0,        # deterministic classification
    max_tokens=20
)

response_text = chat_completion.choices[0].message.content.strip()

# ---------------------------------------------------------------------------
# 4.  Store / display the result
# ---------------------------------------------------------------------------
result = {"response": response_text}
print(result)
```
