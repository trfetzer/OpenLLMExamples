# Example of sending a image prompt with structured JSON response to LLava LLM deployed via llamafile

This is a hands on example that showcases how a user can send an image, here, bergamo.png, to an LLM deployed via llamafile on a local machine.

```python
import base64, json, pathlib, requests

img_b64 = base64.b64encode(pathlib.Path("files/bergamo.png").read_bytes()).decode()

payload = {
    "model": "llava-v1.5-7b-q4",
    "prompt": (
        "USER:[img-12] How many vehicles do you see? "
        "Can you specify their type? For each type can you provide the count in the following JSON schema: "
        "{'type': 'object', 'properties': {"
        "'vehicle_count': { count of number of vehicles in number format }, "
        "'vehicle_types': { Type of vehicle}} \\nASSISTANT:"
    ),
    "image_data": [{"id": 12, "data": img_b64}],
    "id_slot": -1,
    "cache_prompt": False,
    "temperature": 0.1,
    "n_predict": 256
}

resp = requests.post("http://localhost:8080/completion",
                     json=payload)


#print the content of the resp object

print(resp.content)

```
